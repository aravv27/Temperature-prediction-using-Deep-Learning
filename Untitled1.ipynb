```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense,TimeDistributed, RepeatVector,Flatten,Dropout
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import glob
import pandas as pd
from keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
```

```python

training_path = 'D:\\GDG DS\\dataset\\train\\*.csv'
train = glob.glob(training_path)
t_data = pd.concat([pd.read_csv(file) for file in train],ignore_index = True)

valid_path = 'D:\\GDG DS\\dataset\\valid\\*.csv'
valid = glob.glob(valid_path)
v_data = pd.concat([pd.read_csv(vfile) for vfile in valid],ignore_index = True)

```

```python
# print("Training Data Columns:", t_data.columns.tolist())
# print("Validation Data Columns:", v_data.columns.tolist())
print(train_data.shape)
# train_data = t_data.drop('ID',axis = 1,)
# valid_data = v_data.drop('ID',axis = 1)
# train_data = t_data.drop('nx',axis = 1)
# valid_data = v_data.drop('nx',axis = 1)
t_data.columns = t_data.columns.str.strip()
v_data.columns = v_data.columns.str.strip()


train_data = t_data.drop(['ID', 'nx'], axis=1)
valid_data = v_data.drop(['ID', 'nx'], axis=1)
print(train_data.shape)
```

```python

X_train = train_data.iloc[:, :51].values  
y_train = train_data.iloc[:, 51:].values   

X_valid = valid_data.iloc[:, :51].values    
y_valid = valid_data.iloc[:, 51:].values     

X_train_sensors = X_train[:, 1:] 
X_valid_sensors = X_valid[:, 1:] 

X_train = X_train_sensors.reshape(-1, 50, 1)
X_valid = X_valid_sensors.reshape(-1, 50, 1)
```

```python
model = Sequential()

model.add(LSTM(100, activation='relu', input_shape=(50, 1),return_sequences=False))
model.add(Dropout(0.2))
# model.add(LSTM(50, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(2500))
from keras.optimizers import Adam

optimizer = Adam(learning_rate=0.001, clipnorm=1.0)

model.compile(optimizer=optimizer, loss='mse')

early_stopping = EarlyStopping(monitor='val_loss', patience=10,restore_best_weights=True)

model.fit(X_train, y_train, epochs=200, batch_size=25, validation_data=(X_valid, y_valid),callbacks=[early_stopping])

predictions = model.predict(X_valid)  # Predictions on the validation set for evaluation

# Evaluate the model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(y_valid, predictions)
mae = mean_absolute_error(y_valid, predictions)
r2 = r2_score(y_valid, predictions)

print("Validation MSE:", mse)
print("Validation MAE:", mae)
print("Validation R² Score:", r2)
```

```python
model = Sequential()

model.add(LSTM(100, activation='relu', input_shape=(50, 1))) 
model.add(Dense(2500)) 
from keras.optimizers import Adam

optimizer = Adam(learning_rate=0.001, clipnorm=1.0)

model.compile(optimizer=optimizer, loss='mse')

# early_stopping = EarlyStopping(monitor='val_loss', patience=10)

model.fit(X_train, y_train, epochs=175, batch_size=25, validation_data=(X_valid, y_valid))

predictions = model.predict(X_valid)  # Predictions on the validation set for evaluation

# Evaluate the model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(y_valid, predictions)
mae = mean_absolute_error(y_valid, predictions)
r2 = r2_score(y_valid, predictions)

print("Validation MSE:", mse)
print("Validation MAE:", mae)
print("Validation R² Score:", r2)
```

```python
plt.figure(figsize=(14, 7))
plt.plot(y_valid[0], label='Actual', color='blue')
plt.plot(predictions[0], label='Predicted', color='red')
plt.title('Actual vs Predicted Sensor Values')
plt.xlabel('Time Steps')
plt.ylabel('Sensor Values')
plt.legend()
plt.show()
```

```python
error = y_valid[0] - predictions[0]
mae = np.mean(np.abs(error))
rmse = np.sqrt(np.mean(error**2))

# Create a figure with subplots
plt.figure(figsize=(14, 7))

# Plot actual values
plt.plot(y_valid[0], label='Actual', color='blue', linewidth=2)

# Plot predicted values
plt.plot(predictions[0], label='Predicted', color='red', linewidth=2)

# Highlight the error
plt.fill_between(range(len(error)), y_valid[0], predictions[0], color='gray', alpha=0.3, label='Error Area')

# Adding error metrics to the title
plt.title(f'Actual vs Predicted Sensor Values\nMAE: {mae:.4f}, RMSE: {rmse:.4f}', fontsize=14)
plt.xlabel('Time Steps', fontsize=12)
plt.ylabel('Sensor Values', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
